```python
import tensorflow as tf
from tensorflow import keras

# Data Preparation
(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# dng v  th, hnh nh
import matplotlib.pyplot as plt 
import numpy as np

# To dang sch 9 phn t ngu nhin t m_train c 60000 phn t
indices = list(np.random.randint(60000, size=64))

fig = plt.figure(figsize=(9,9))
columns = 8
rows = 8
for i in range(1, columns*rows +1):
    img = X_train[indices[i-1]].reshape(28,28)
    fig.add_subplot(rows, columns, i)
    
    plt.axis('off')
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
import tensorflow as tf
from tensorflow import keras

# Data Preparation
(X_train, y_train), (X_test, y_test) = keras.datasets.cifar10.load_data()

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

fig =plt.figure(figsize=(12,12))

for i in range(10):
    X_train_i = X_train[y_train[:,0]==i]    
    X_train_i = X_train_i[:10]
    
    for j in range(10):
        img = X_train_i[j].reshape(32,32,3)
        fig.add_subplot(10, 10, i*10 + j+1)
        plt.axis('off')
        plt.imshow(img, vmin=0, vmax=255)
```
